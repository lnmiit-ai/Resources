```python
#STEP1
import numpy as np

data=np.loadtxt('/home/aasim/Resources/Week 1 /assignment/data.txt',delimiter='\t',skiprows=1)
print(data.shape)
print(type(data))
m,n=data.shape
print(m,n)
print(data[0,1:])
```

```python
#Sorting rows according to label 1 and 2 and storing them in indivdual numpy matrices
label1=np.zeros((1,n-1)).reshape((1,n-1))#1x10
label2=np.zeros((1,n-1)).reshape((1,n-1))#1x10

for i in range(0,m):
    if(data[i,0]%2==0):#i.e. if label=2
        label2=np.concatenate((label2,data[i,1:].reshape((1,n-1))))
    else:
        label1=np.concatenate((label1,data[i,1:].reshape((1,n-1))))

label1=label1[1:,:]
label2=label2[1:,:]
print('Label 1:',label1.shape)
print('Label 2:',label2.shape)
m1,n1=label1.shape
m2,n2=label2.shape
```

```python
#STEP2
import matplotlib.pyplot as plt

pltcnt=0
for i in range(0,n1):
    for j in range(i+1,n1):
        plt.title('Feature '+str(j+1)+' vs Feature '+str(i+1))
        plt.xlabel('Feature '+str(i+1))
        plt.ylabel('Feature '+str(j+1))
        plt.scatter(label1[:,i],label1[:,j],color='red',marker='.')
        plt.scatter(label2[:,i],label2[:,j],color='blue',marker='.')
        plt.legend(['Label 1','Label 2'])
        plt.show()
```
