```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data.txt',skiprows=1)
```

```python
d1 = np.array([x for x in data if x[0] == 1 ]) #to separate two different labels in the dataset
d2 = np.array([x for x in data if x[0] == 2 ])
```

```python
for i in range(1,11):
    for j in range(i+1,11):
        plt.title("Scatter plot of data taking two features")       
        plt.xlabel("Feature" + " " + str(i))
        plt.ylabel("Feature" + " " + str(j))
        plt.scatter(d1[:,i],d1[:,j],c='r') #(x,y,color)
        plt.scatter(d2[:,i],d2[:,j],c='b')
        plt.legend(['Label '+str(i), 'Label '+str(j)])
        plt.show()
```

# By Visualising the scatter plots, only feature 1 and feature 2 can be clearly distinguished
