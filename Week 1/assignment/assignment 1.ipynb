```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('data.txt',skiprows=1)
```

```python
data
```

```python
d1=np.array([x for x in data if x[0]==1])
```

```python
d1
```

```python
d2=np.array([x for x in data if x[0]==2])
```

```python
d2
```

```python
%matplotlib inline
```

```python
for i in range(11):
    for j in range(i+1,11):
        plt.title('Plot of two features')
        a=d1[:,i]
        b=d1[:,j]
        c=d2[:,i]
        d=d2[:,j]
        plt.plot(a,b,'b*',c,d,'r*')
        plt.show()
```
