```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
defa=np.loadtxt('data.txt', skiprows=1)
```

```python
defa.shape
```

```python
#Here we separte data having on basis of label at zeroth column
lab1= np.array([x for x in defa if x[0] == 1 ])
lab2= np.array([x for x in defa if x[0] == 2 ])
```

```python
lab1.shape
```

```python
lab2.shape
```

```python
for i in range(1,11):
    for j in range(i+1,11):
        plot.title("Scatter plot of data taking two features")       
        plot.xlabel("Feature" + " " + str(i))
        plot.ylabel("Feature" + " " + str(j))
        plot.scatter(lab1[:,i],lab1[:,j],c='r') #(x,y,color)
        plot.scatter(lab2[:,i],lab2[:,j],c='b')
        plot.legend(['Label 1', 'Label 2'])
        plot.show()
```
